```python
#On se concentre sur les produits tag France/fr
french= df_full.loc[df_full['countries'].str.contains('fr',na=False, case=False)]
```

```python
print(f"Taille du df total: {len(french)}")
print(f"% de donnée minimum: {0.20}")
print(f"Taille minimum de données acceptée: {len(french)*0.20}")
```

```python
french.isnull().sum().sort_values(ascending = False)[french.isnull().sum().sort_values(ascending = False)/len(french)>0.8]
```

```python
#pourcentage de valeurs non-null
percent= 0.20
#calcul du Threshold (Nombre de valeurs réel)
tresh= len(french)*percent
french2= french.dropna(axis=1,thresh=tresh, subset=None, inplace=False)
```

```python
#premier dataset nettoyé d'une bonne partie des colonnes à valeurs nulls
french2.head()
```

```python
#french2.to_csv("french.csv", index = False) # <= df servant pour la partie notebook_sql
```

```python
french2['proteins_100g'].isnull().sum()
```

```python
#gérer les valeurs nulls en les transformant en 0
value = {"fat_100g": 0,
"sugars_100g" : 0,
"proteins_100g": 0,
"salt_100g": 0,
"energy-kcal_100g": 0,
"energy_100g": 0}
df_fill = french2.fillna(value)
```

```python
df_fill['proteins_100g']
```

```python
#drop des colonnes inutiles
labels = df_fill[['image_url','image_small_url',\
                  'image_ingredients_url','image_ingredients_small_url',\
                  'image_nutrition_url','image_nutrition_small_url']]

df = df_fill.drop(labels, axis=1)
```

```python
df.shape
```

```python
df.info()
```

```python
#Delete des produits avec proteins = 0
df_proteins = df[df.proteins_100g != 0]
```

```python
df_proteins = df_proteins[df_proteins.fat_100g != 0]
```

```python
df_proteins = df_proteins[df_proteins.sugars_100g != 0]
```

```python
df_proteins = df_proteins[df_proteins.salt_100g != 0]
```

```python
df_proteins = df_proteins[df_proteins.energy_100g != 0]
```
